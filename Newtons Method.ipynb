In this notebook we are going to be exploring how to implement newtons method. We will start with an explanation of the functions used and then we will provide many examples of the function in use so that we can gain a better intuition for this method before moving on to more complex topics, such as Runge-Katta methods

We will start simply with importing the necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#In this cell we define the function thats on the other side of the equality from the derivative.
#This can be whatever we want aslong as it stays a function of x and y.
#In this case, since we need the function to be filled with something, our function is
#f(x,y)=x+(1/5)y

def functionInQuestion(x,y):
    return(x+((1/5)*y))
```

```python
#Generalized Function that takes in the stepsize, initial conditions and said function from previous cell and returns
#a 2d array where the first entry is the array of x-values and the second entry is the array of approximated y-values

def newtonsMethod(stepSize, initialPair, function):
    y0=initialPair[1]
    x0=initialPair[0]
    deltat=stepSize

    lasty = y0
    lastx = x0

    solutionx = np.array([x0])
    solutiony = np.array([y0])
    iterations = math.floor(6/deltat)
    for i in range(iterations):
        nexty = lasty+(deltat*(function(lastx,lasty)))
        nextx = lastx+deltat
        solutionx = np.append(solutionx, [nextx],axis=0)
        solutiony = np.append(solutiony, [nexty],axis=0)
        lasty=nexty
        lastx+=deltat
        
    solution = np.array([solutionx, solutiony])
    return(solution)
```

```python
#And in this third and last cell we simply plot that 2D array that we got from newtonsMethod

solution = newtonsMethod(1,np.array([0,-3]),functionInQuestion)
plt.plot(solution[0], solution[1])
```

Now that we have shown what our functions are and how to implement them, we will now use them with many other functions so we can explore what Newtons Method holds for us

```python
def functionInQuestion(x,y):
    return(x+((1/5)*y))

solution1 = newtonsMethod(1,np.array([0,-3]),functionInQuestion)
solution2 = newtonsMethod(0.5,np.array([0,-3]),functionInQuestion)
solution3 = newtonsMethod(0.11,np.array([0,-3]),functionInQuestion)
plt.plot(solution1[0], solution1[1], label="Step Size = 1")
plt.plot(solution2[0], solution2[1], label="Step Size = 0.5")
plt.plot(solution3[0], solution3[1], label="Step Size = 0.1")
plt.legend()
```

```python
def functionInQuestion(x,y):
    return(32-0.16*y)

initialConditions = np.array([0,0])

solution1 = newtonsMethod(1,initialConditions,functionInQuestion)
solution2 = newtonsMethod(0.5,initialConditions,functionInQuestion)
solution3 = newtonsMethod(0.2,initialConditions,functionInQuestion)
solution4 = newtonsMethod(0.1,initialConditions,functionInQuestion)
print(solution2[1])
plt.plot(solution1[0], solution1[1], label="Step Size = 1")
plt.plot(solution2[0], solution2[1], label="Step Size = 0.5")
plt.plot(solution3[0], solution3[1], label="Step Size = 0.2")
plt.plot(solution4[0], solution4[1], label="Step Size = 0.1")
plt.legend()
```
