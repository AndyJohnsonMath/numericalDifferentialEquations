```python
def newtonsMethod(stepSize, initialPair, function):
    y0=initialPair[1]
    x0=initialPair[0]
    deltat=stepsize

    lasty = y0
    lastx = x0

    solutionx = np.array([x0])
    solutiony = np.array([y0])
    for i in range(6):
        nexty = lasty+(deltat*(lastx+((0.2)*lasty)))
        nextx = lastx+deltat
        solutionx = np.append(solutionx, [nextx],axis=0)
        solutiony = np.append(solutiony, [nexty],axis=0)
        lasty=nexty
        lastx+=deltat
        
    solution = np.array([solutionx, solutiony])
    return(solution)
```
